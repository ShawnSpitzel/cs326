```python
def decode_utf8_bytes_to_str_wrong(bytestring: bytes):
    ls = []
    print(bytestring)
    for b in bytestring:
        print(f"byte: {b}, char: {bytes([b]).decode('utf-8')}")
        ls.append(bytes([b]).decode("utf-8"))
    return "".join(ls)
```
